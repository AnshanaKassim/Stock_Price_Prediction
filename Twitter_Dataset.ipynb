# **Extraction of Apple Stocks Tweets**

### Importing Libraries

```python
import pandas as pd
import numpy as np
```

```python
twitter_df = pd.read_csv('Tweet.csv')
twitter_df
```

```python
# Extract Apple stock tweets from the Dataset
company_tweets = pd.read_csv('Company_Tweet.csv')
aapl_ticker_ids = company_tweets[company_tweets['ticker_symbol'] == 'AAPL']['tweet_id'].unique()
if len(aapl_ticker_ids) == 0:
    print("No data found for AAPL.")
else:
    print("Tweet IDs for AAPL:", aapl_ticker_ids)
    tweets = pd.read_csv('Tweet.csv')
    aapl_tweets = tweets[tweets['tweet_id'].isin(aapl_ticker_ids)]
    print(aapl_tweets.shape)
```

```python
# Convert the 'post_date' column to datetime format
aapl_tweets.loc[:, 'post_date'] = pd.to_datetime(aapl_tweets['post_date'], unit='s')
```

```python
# Save Apple tweets
aapl_tweets.to_csv('AAPL_Tweets.csv', index=False)
```
